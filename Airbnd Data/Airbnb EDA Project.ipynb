**Task 4 : Data Cleaning**

```python
df.isnull().sum()
```

```python
# Dropping All Missing Values Rows
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# Dealing with duplicates rows
df.duplicated().sum()
# Deleting all duplicates rows
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
# Type casting 
# Changing Data types
df.dtypes
df['id'] = df['id'].astype(object)
df['host_id'] = df['host_id'].astype(object)
df.dtypes
```

**Task 5 : Data Analysis**

```python
# Identfying outliers in price
data = df[df['price'] < 1500 ]
sns.boxplot(data=data , x='price')
plt.show()
```

```python
# Price Distribution 
plt.figure(figsize=(8,5))
sns.histplot(data=data , x= 'price' ,bins=100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
# Availability_365  Distribution
plt.figure(figsize=(6,3))
sns.histplot(data=data ,x='availability_365')
plt.title("Availability_365 Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create Price Per bed column
data['price per bed'] = data['price']/data['beds']
```

```python
data.head()
```

```python
# Average Price Per beds
data.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
df.columns
```

```python
# Price Dependency on neighbourhood_group
sns.barplot(data=data , x='neighbourhood_group',y= 'price',hue='room_type')
plt.title('Price Dependency on Neighbourhood Group By Room Type')
plt.show()
```

```python
# Number of reviews And Price reletade
plt.figure(figsize=(8,5))
sns.scatterplot(data=data,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.title("Locality and review Dependency")
plt.show()
```

```python
data.dtypes
```

```python
sns.pairplot(data=data,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue= 'room_type')
plt.show()
```

```python
# Geographical Distribution Of AirBnb Listing
plt.figure(figsize=(10,7))
sns.scatterplot(data=data,x='longitude' , y='latitude' , hue= 'room_type')
plt.title('Geographical Distribution Of AirBnb Listing')
plt.show()
```

```python
# Heat map - correlation of one variable with others for numerical column
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
plt.figure(figsize=(12,8))
sns.heatmap(data=corr ,annot= True ,cmap='Greens')
plt.show()
```
